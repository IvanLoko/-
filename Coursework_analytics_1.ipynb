```python
dates = []
current_date = np.datetime64('2016-03')
while current_date != np.datetime64('2019-08'):
    dates.append(str(current_date))
    current_date = current_date + np.timedelta64(1,'M')
# empty_df = pd.DataFrame(columns=dates, index=main.title.unique())
```

```python
df_student_growth = tot_courses[tot_courses.is_homework==True].groupby(['title']).\
agg({'module_number':'first', 'lesson_number':'first'}).reset_index()
df_student_growth = df_student_growth.merge(main)
df_student_growth.start_date = df_student_growth.start_date.transform(lambda x: str(x)[0:7])

df_student_growth = df_student_growth.groupby(['title', 'start_date']).count().\
reset_index()[['title', 'student_id', 'start_date']]
df = pd.DataFrame({'title' : sorted(list(df_student_growth.title.unique())*len(dates), key=lambda x: x[0:]),
                   'start_date' : dates*15})

df_student_growth = df.merge(df_student_growth, on=['title', 'start_date'], how='outer').fillna(0)
```

Почти у всех курсов имеется локальный пик, приходящийся на период с декабрь 2018 по январь 2019 года, логично предположить, что после резкого прироста, последущие значения будут меньше. Однако, курсы "Java разработчик", "Анимация интерфейсов", "Веб-верстка для начинающий 2.0" имеют почти нулевой показатель прироста студентов за последние месяцы. Курсы "Веб-дизан Базовый", "Веб-разработчик" находятся в схожей ситуации, но, с более ранним пиком. 

```python
def date(row):
    if pd.isna(row.finish_date) and row.status!='fail':
        row.finish_date = main.finish_date.max().date()
    elif row.status=='fail':
        row.finish_date = row.start_date
    else:
        row.finish_date = row.finish_date.date()
    return row
```

```python
def ym(row):
    current_date = row.start_date
    ym = [[current_date.year, current_date.month]]
    while current_date.year != row.finish_date.year or current_date.month != row.finish_date.month:
        current_date += relativedelta(months=1)
        ym.append([current_date.year, current_date.month])
    return ym
```

```python
homework_df = main[(main.is_homework == True)].drop(main.columns[1:12], axis=1)
```

```python
homework_df = homework_df.apply(date, axis=1)
homework_df.start_date = homework_df.start_date.transform(lambda x: x.date())
homework_df['yearMonth'] = homework_df.apply(ym, axis=1).transform(lambda x: str(x))
```

```python
df_hw_count = pd.DataFrame(columns = dates, index=courses_list)
current_date = dt.date(2016,3,1)
column_date = np.datetime64('2016-03')
while current_date != dt.date(2019,8,1):
    date = f'\[{current_date.year}, {current_date.month}\]'
    df_hw_count[str(column_date)] = homework_df[homework_df.yearMonth.str.contains(date)].groupby('title').count()
    current_date += relativedelta(months=1)
    column_date += np.timedelta64(1, 'M')
df_hw_count.fillna(0, inplace=True)
```

```python
fig, ax = plt.subplots(5,3, figsize=(20,20))
plt.subplots_adjust(wspace=0.1, hspace=0.5)
plt.suptitle('Нагрузка на преподавателей', fontsize = 20, y=0.93)
i=0
for row in range(5):
    for column in range(3):
        ax[row, column].plot(df_hw_count.columns, df_hw_count.loc[courses_list[i]] , label = courses_list[i])
        ax[row, column].set_title(courses_list[i])
        ax[row, column].grid(alpha=0.5)
        ax[row, column].set_xticks([df_hw_count.columns[i] for i in range(0, len(df_hw_count.columns)+1, 2)])
        ax[row, column].set_xticklabels([df_hw_count.columns[i] for i in range(0, len(df_hw_count.columns)+1, 2)], 
                                            rotation=45)

        i+=1
plt.setp(ax[0:, 0], ylabel='Количество актуальных\nдомашних заданий')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,10))
plt.title('Нагрузка на преподавателей', fontsize=20)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.patch.set_facecolor('grey')
ax.patch.set_alpha(0.02)
ax.grid(alpha=0.4, c='grey', linestyle='--')
for course in courses_list:
    ax.plot(df_hw_count.columns, df_hw_count.loc[course] , label = course, linewidth=0.9)
plt.legend(loc='upper left', fontsize=16, framealpha=0)
ax.set_xticks(dates)
ax.set_xticklabels(dates, rotation=45)
plt.ylabel('Количество новых студентов', fontsize=14)
plt.show()
```

При сопоставлении данных о ежемесячном приросте и нагрузке преподавателей, видно, что количество актуальных домашних заданий растет соответсвенно приросту студентов на курсе.
